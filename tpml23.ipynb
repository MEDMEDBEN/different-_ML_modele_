```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

X = df[['Humidity','Loud Cover','Temperature (C)']]  
y = df['Visibility (km)']  

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Erreur quadratique moyenne (MSE):", mse)
print("Coefficient de détermination (R²):", r2)
print("Coefficients :", model.coef_)
print("Intercept :", model.intercept_)


plt.figure(figsize=(6, 6))
plt.scatter(y_test, y_pred, color='blue', alpha=0.7)
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linestyle='--')
plt.xlabel("Valeurs réelles")
plt.ylabel("Valeurs prédites")
plt.title("Régression linéaire : Réel vs Prédit")
plt.show()

residus = y_test - y_pred
plt.figure(figsize=(6, 4))
plt.scatter(y_pred, residus, color='purple', alpha=0.7)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Valeurs prédites")
plt.ylabel("Résidus (erreurs)")
plt.title("Analyse des résidus")
plt.show()

plt.figure(figsize=(7, 4))
plt.bar(range(len(model.coef_)), model.coef_)
plt.xticks(range(len(X.columns)), X.columns, rotation=45)
plt.title("Importance des variables (coefficients du modèle)")
plt.show()
```
